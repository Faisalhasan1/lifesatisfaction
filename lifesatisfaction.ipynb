```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import sklearn
```

```python
oecd_bli=pd.read_csv("C:\oecdml\BLI_24.csv",thousands=',')
gpd_per_capita=pd.read_csv("C:\oecdml\WEO_Data.csv",thousands=',',delimiter='\t',encoding='latin1',na_values="n/a")
```
